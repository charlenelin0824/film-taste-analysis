```python
# I want to know how many people endorsed one of my favorite films: Yi Yi, and where they're from
YiYi = df[df['Movie'].str.contains('Yi Yi', case=False, na=False)]
YiYi['Critic Country'].value_counts()
```

```python
# Votes for voted movie in each country
df1 = df.groupby('Critic Country')['Movie'].value_counts().sort_values(ascending=False).reset_index(name='count')
df1.head(60)
```

```python
# Votes for voted movie, sorted by counts
df2 = df['Movie'].value_counts().reset_index(name='count')
df2.head(60)
```
