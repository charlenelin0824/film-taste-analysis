## Search on TMDB -- Test

```python
import requests
import urllib.parse

def search_TMDb_and_get_page1(movie_name):
    # url encode the movie name using urllib.parse.quote
    movie_name = urllib.parse.quote(movie_name)

    # create the url
    url = f"https://api.themoviedb.org/3/search/movie?query={movie_name}&include_adult=false&language=en-US&page=1"

    # hit api
    response = requests.get(url, headers=headers)

    # return the response
    return response.json()['results']

```

```python
movie_name = "In the Mood for Love"
search_results = search_TMDb_and_get_page1(movie_name)
# turn it into a dataframe
search_results_df = pd.DataFrame(search_results)
search_results_df['BBC Name'] = movie_name
search_results_df
```

## Search all movies on BBC list

```python
from time import sleep

all_search_results = []

# loop through BBC
for BBC_name in BBC['Movie']:
    print(BBC_name)
    search_results = search_TMDb_and_get_page1(BBC_name)
    print(f"found {len(search_results)} results")
    print('---')
    sleep(0.1) # sleep for 0.1 seconds
    
    all_search_results.append(search_results_df)
```

```python
import pandas as pd

# If all_search_results is a list of pandas DataFrames or Series
all_search_results_df = pd.concat(all_search_results, ignore_index=True)

# If you need to reset the index (and drop the old one)
all_search_results_df = all_search_results_df.reset_index(drop=True)

all_search_results_df
```

```python
# Loop through each unique BBC Name, keep the result with the highest popularity
final_results = []
for BBC_name in all_search_results_df['BBC Name'].unique():
    # get the rows for this BBC name
    rows = all_search_results_df[all_search_results_df['BBC Name'] == BBC_name]
    # sort the rows by popularity
    rows = rows.sort_values('popularity', ascending=False)
    # take the top row
    top_row = rows.iloc[0]
    # add it to the final results
    final_results.append(top_row)

final_results_df = pd.DataFrame(final_results)
final_results_df
```

```python
# save to csv
all_search_results_df.to_csv('TMDb_search_results.csv', index=False)
```

## Find IMDb ID from IMDb title.basics.tsv

```python
# read tsv file "title.basics.tsv" into a dataframe
title_basics = pd.read_csv('title.basics.tsv', sep='\t')
```

```python
title_basics
```

```python
# Loop through BBC list and get the IMDb ID
IMDb_ids = []
for BBC_name in BBC['Movie']:
    print(BBC_name)
    # get the row for this BBC name
    row = title_basics[title_basics['primaryTitle'] == BBC_name]
    # get the IMDb ID
    IMDb_id = row['tconst']
    # add it to the list
    IMDb_ids.append(IMDb_id)
```

```python
# Make a new dataframe with BBC names and their IMDb IDs 
BBC_IMDb = pd.DataFrame({'BBC Name': BBC['Movie'], 'IMDb ID': IMDb_ids})
BBC_IMDb
```

## Try out OMDb API

```python
# You start knowing nothing
results = []

# Loop through each of your movie names
for movie_name in BBC['Movie']:
    # Ask OMDb for the information
    # this is the code you have right now
    # http://www.omdbapi.com/?t=fast+and+furious
    movie_name = movie_name.replace(' ', '+')
    url = f"http://www.omdbapi.com/?t={movie_name}&apikey={OMDB_API_KEY}"
    result = requests.get(url).json()
    # Take the information and save it
    results.append(result)
results
```
